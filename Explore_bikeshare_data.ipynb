# Question 1


**What are the counts of each user type?**

```r
## Function to get counts of each user type
## Convert to table to get counts and convert back to dataframe
## remove blank space user types
usertype_counts = function(dataset, column) {
  column_counts = table(dataset[[column]])
  column_counts = as.data.frame(column_counts)
  colnames(column_counts) = c(column, "Counts")
  column_counts = subset(column_counts, column_counts[[column]] != "")
  return(column_counts)
}
```

### New York

```r
unique(ny$User.Type)
```

```r
## Use usertype_counts function
usertype_ny = usertype_counts(ny, "User.Type")
usertype_ny
```

```r
ggplot(data = usertype_ny, aes(x = User.Type, y = Counts, fill = User.Type)) +
  geom_bar(stat = "identity") +
  labs(title = "User Type Counts",
       x = "User Types",
       y = "Counts")
```

# Question 3

**What is the average travel time by user type?**

### New York

```r
ad_ny = aggregate(Trip.Duration ~ User.Type, data = ny, FUN = mean)
colnames(ad_ny) <- c("User.Type", "Average.Trip.Duration")
ad_ny = subset(ad_ny, User.Type != "")
```

```r
print(ad_ny)
```

```r
ggplot(data = ad_ny, aes(x = User.Type, y = Average.Trip.Duration, fill = User.Type)) +
  geom_bar(stat = "identity") +
  labs(title = "Average Trip Duration by User Type",
       x = "User Type",
       y = "Average Trip Duration")
```

### Washington

```r
ad_wash = aggregate(Trip.Duration ~ User.Type, data = wash, FUN = mean)
colnames(ad_wash) <- c("User.Type", "Average.Trip.Duration")
ad_wash = subset(ad_wash, User.Type != "")
```

```r
print(ad_wash)
```

```r
ggplot(data = ad_wash, aes(x = User.Type, y = Average.Trip.Duration, fill = User.Type)) +
  geom_bar(stat = "identity") +
  labs(title = "Average Trip Duration by User Type",
       x = "User Type",
       y = "Average Trip Duration")
```

### Chicago

```r
ad_chi = aggregate(chi$Trip.Duration, by = list(chi$User.Type), FUN = mean)
colnames(ad_chi) <- c("User.Type", "Average.Trip.Duration")
ad_chi = subset(ad_chi, User.Type != "")
```

```r
print(ad_chi)
```

```r
ggplot(data = ad_chi, aes(x = User.Type, y = Average.Trip.Duration, fill = User.Type)) +
  geom_bar(stat = "identity") +
  labs(title = "Average Trip Duration by User Type",
       x = "User Type",
       y = "Average Trip Duration")
```

### Average Trip Duration by User Type:

**New York:**
- Customer: 2193.1 seconds/36.6 minutes
- Subscriber: 755.4 seconds/12.6 minutes

**Washington: Columbus Station/Union Circle**
- Customer: 2634.4 seconds/43.9 minutes
- Subscriber: 733.3 seconds/12.2 minutes

**Chicago: Streeter Dr & Grand Ave**
- Customer: 1930.0 seconds/32.2 minutes

**In all three cities, customers on average took longer trips than subscribers.**
